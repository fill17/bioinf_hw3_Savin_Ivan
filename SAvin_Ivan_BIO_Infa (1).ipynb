**1. Скачивание индивидуальных файлов**
- `.hisat` файлы - статистика выравнивания из HISAT2
- `.counts` файлы - количество чтений на каждый ген
- FastQC отчеты - HTML/ZIP файлы с анализом качества данных

**2. Создание ALL.counts**
- все индивидуальные `.counts` файлы в одну таблицу
- Переименовывает образцы: SRR3414629 → r1, SRR3414635 → c1 и т.д.
- Сохраняет в формате TSV (таб-разделитель)

**3. Генерация таблицы статистики**
- Собирает ключевую статистику из HISAT2 и HTSeq-count

**4. Создание MultiQC отчета**
- Генерирует объединенный отчет по качеству всех образцов

- `ALL.counts` - основная таблица для DESeq2 анализа
- `alignment_statistics.csv` - статистика выравнивания
- `*.hisat` - детальная статистика по каждому образцу
- `*.counts` - индивидуальные counts файлы
- FastQC отчеты - визуализация качества данных


```python
from google.colab import files
import pandas as pd
import os

samples = {
    'SRR3414629': 'r1',
    'SRR3414630': 'r2',
    'SRR3414631': 'r3',
    'SRR3414635': 'c1',
    'SRR3414636': 'c2',
    'SRR3414637': 'c3'
}
for sra_id, short_name in samples.items():
    print(f"\nОбрабатываем {sra_id} ({short_name})...")

    hisat_file = f"{sra_id}.hisat"
    if os.path.exists(hisat_file):
        files.download(hisat_file)
        print(f"  {hisat_file} - статистика HISAT2")

    counts_file = f"{sra_id}.counts"
    if os.path.exists(counts_file):
        files.download(counts_file)
        print(f"  {counts_file} - counts по генам")

    fastqc_html = f"{sra_id}_1_fastqc.html"
    fastqc_zip = f"{sra_id}_1_fastqc.zip"
    if os.path.exists(fastqc_html):
        files.download(fastqc_html)
        print(f"  {fastqc_html} - FastQC отчет")
    elif os.path.exists(fastqc_zip):
        files.download(fastqc_zip)
        print(f"  {fastqc_zip} - FastQC отчет (zip)")

def create_all_counts():
    print("Создаем ALL.counts...")

    all_data = {}
    gene_ids = None

    for sra_id, short_name in samples.items():
        counts_file = f"{sra_id}.counts"

        if os.path.exists(counts_file):
            print(f"Читаем {counts_file} -> {short_name}")

            counts_dict = {}
            with open(counts_file, 'r') as f:
                for line in f:
                    if not line.startswith('__'):
                        parts = line.strip().split('\t')
                        if len(parts) == 2:
                            gene, count = parts
                            counts_dict[gene] = int(count)

            all_data[short_name] = counts_dict

            if gene_ids is None:
                gene_ids = list(counts_dict.keys())
        else:
            print(f"{counts_file} не найден")

    if gene_ids is not None:
        df = pd.DataFrame({'gene_id': gene_ids})

        for short_name in ['c1', 'c2', 'c3', 'r1', 'r2', 'r3']:
            if short_name in all_data:
                df[short_name] = [all_data[short_name].get(gene, 0) for gene in gene_ids]

        df.to_csv('ALL.counts', sep='\t', index=False)
        files.download('ALL.counts')
        print("ALL.counts создан и скачан!")

        return df
    else:
        print("Не найдено ни одного .counts файла!")
        return None

all_counts_df = create_all_counts()

def create_alignment_stats():


    stats_data = []

    for sra_id, short_name in samples.items():
        hisat_file = f"{sra_id}.hisat"
        counts_file = f"{sra_id}.counts"

        if not os.path.exists(hisat_file):
            continue

        stats = {
            'Sample_ID': sra_id,
            'Short_Name': short_name,
            'Type': 'Reprogrammed' if short_name.startswith('r') else 'Control'
        }

        with open(hisat_file, 'r') as f:
            hisat_content = f.read()

        for line in hisat_content.split('\n'):
            if 'reads; of these:' in line:
                stats['Total_Reads'] = int(line.split()[0])
            elif 'aligned exactly 1 time' in line:
                stats['Uniquely_Mapped'] = int(line.split()[0])
                stats['Uniquely_Mapped_Percent'] = float(line.split('(')[1].split('%')[0])
            elif 'aligned >1 times' in line:
                stats['Multi_Mapped'] = int(line.split()[0])
            elif 'overall alignment rate' in line:
                stats['Overall_Alignment_Rate'] = float(line.split('%')[0])

        if os.path.exists(counts_file):
            with open(counts_file, 'r') as f:
                counts_lines = f.readlines()

            for line in counts_lines:
                if line.startswith('__no_feature'):
                    stats['No_Feature'] = int(line.split('\t')[1])
                elif line.startswith('__ambiguous'):
                    stats['Ambiguous'] = int(line.split('\t')[1])
                elif line.startswith('__too_low_aQual'):
                    stats['Too_Low_Quality'] = int(line.split('\t')[1])
                elif line.startswith('__not_aligned'):
                    stats['Not_Aligned'] = int(line.split('\t')[1])
                elif line.startswith('__alignment_not_unique'):
                    stats['Alignment_Not_Unique'] = int(line.split('\t')[1])

            total_gene_reads = 0
            for line in counts_lines:
                if not line.startswith('__'):
                    total_gene_reads += int(line.split('\t')[1])
            stats['Total_Gene_Reads'] = total_gene_reads

        stats_data.append(stats)

    if stats_data:
        stats_df = pd.DataFrame(stats_data)

        stats_df.to_csv('alignment_statistics.csv', index=False)
        files.download('alignment_statistics.csv')


        display_cols = ['Short_Name', 'Type', 'Total_Reads', 'Uniquely_Mapped',
                       'Uniquely_Mapped_Percent', 'Total_Gene_Reads']
        available_cols = [col for col in display_cols if col in stats_df.columns]

        print(stats_df[available_cols].to_string(index=False))

        return stats_df
    else:
        print("Не удалось собрать статистику")
        return None

stats_df = create_alignment_stats()

print("\n" + "="*50)
print("4. MULTIQC ОТЧЕТ")
print("="*50)

try:
    !pip install multiqc > /dev/null 2>&1
    !multiqc . -o multiqc_results/ > /dev/null 2>&1
    files.download('multiqc_results/multiqc_report.html')
    print("MultiQC отчет создан и скачан!")
except:
    print("MultiQC не установлен или произошла ошибка")
```
